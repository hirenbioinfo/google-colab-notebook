To run R code on Google Colab, follow these steps:

   - Install any required R packages using `install.packages()` within an R code cell.

2. Start an R code cell:
   - Use `%%R` at the beginning of the cell to specify that the code is written in R.

3. Write and execute R code:
   - Write your R code within the R code cell and run the cell.
   - The output and plots generated by the R code will be displayed in the output area below the cell.
